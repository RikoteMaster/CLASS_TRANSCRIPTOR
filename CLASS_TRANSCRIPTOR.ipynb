```python
from transformers import pipeline
import torch

# Configurar la pipeline de reconocimiento de voz
device = "cuda:0" if torch.cuda.is_available() else "cpu"
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-large-v2",
    chunk_length_s=30,
    device=device,
)


```

```python

archivo = 'prueba_audio.mp3'
print(f"Procesando {archivo}...")
prediction = pipe(archivo, batch_size=8, return_timestamps=False)

print(prediction)
# Guardar las predicciones en un archivo de texto


print("Predicciones guardadas en 'predicciones.txt'")
```

```python
with open('predicciones.txt', 'w', encoding='utf-8') as file:
    # Escribe cada predicciรณn en el archivo
        file.write(str(prediction['text']) + '\n')
```
